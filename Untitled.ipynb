```python
import os
import numpy as np
import pandas as pd
from itertools import combinations, groupby
from collections import Counter
```

```python
movies = pd.read_csv('ml-20m/movies.csv')
ratings = pd.read_csv('ml-20m/ratings.csv')
```

```python
ratings
```

```python
ratings["timestamp"] = pd.to_datetime(ratings['timestamp'], unit='s')
```

```python
valid_date = (2000 < ratings["timestamp"].dt.year) & (ratings["timestamp"].dt.year < 2019) & (ratings["rating"] > 4.0)
ratings[valid_date].shape
```

```python
ratings[~valid_date].shape
```

```python
trs = {}
for i, row in ratings[valid_date][['userId','movieId']].iterrows():
    try:
        trs[row["userId"]].append(row["movieId"])
    except KeyError:
        trs[row["userId"]] = []
        trs[row["userId"]].append(row["movieId"])
    
```

```python
from efficient_apriori import apriori
itemsets, rules = apriori(trs.values(), min_support=0.01,  min_confidence=0.5, max_length=2)
```

```python
print(rules)  # [{eggs} -> {bacon}, {soup} -> {bacon}]
rules=sorted(rules, key=lambda rule: rule.confidence)
for rule in rules:
  print(rule) # Prints the rule and its confidence, support, lift, ...
```

```python
rules=sorted(rules, key=lambda rule: rule.confidence)
for rule in rules:
    print("{%s} -> {%s} (conf: %.3f, supp: %.3f, lift: %.3f, conv %.3f)\n\n" % \
          (movies[movies["movieId"] == rule.lhs[0]].title.values[0],
           movies[movies["movieId"] == rule.rhs[0]].title.values[0],
           rule.confidence, rule.support, rule.lift, rule.conviction
          )
         ) # Prints the rule and its confidence, support, lift, ...
```
